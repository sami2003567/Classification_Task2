```python
# Imports
import os, sys, warnings, pickle, textwrap
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path
from typing import Tuple, Dict

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score, roc_curve, auc, accuracy_score
)
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.feature_selection import RFECV

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB

# Try SMOTE if available
try:
    from imblearn.over_sampling import SMOTE
    HAS_SMOTE = True
except Exception:
    HAS_SMOTE = False

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

print("SMOTE available:", HAS_SMOTE)
```

## 1) Load Data
Tries the following, in order:
- Local `mushrooms.csv` (preferred when offline).
- `ucimlrepo` to fetch the **Agaricus-Lepiota Mushroom** dataset (requires internet).

```python
def load_mushroom_data() -> pd.DataFrame:
    # 1) local CSV
    possible_paths = [
        Path('mushrooms.csv'),
        Path('/mnt/data/mushrooms.csv'),
        Path.cwd() / 'mushrooms.csv'
    ]
    for p in possible_paths:
        if p.exists():
            print(f"Loading local CSV: {p}")
            return pd.read_csv(p)
    # 2) ucimlrepo fetch (if available & online)
    try:
        from ucimlrepo import fetch_ucirepo
        print("Attempting to fetch via ucimlrepo...")
        # UCI Mushroom dataset ID is 73 in ucimlrepo (Agaricus-Lepiota)
        mushroom = fetch_ucirepo(id=73)
        df = pd.DataFrame(mushroom.data.features)
        df['class'] = mushroom.data.targets.squeeze()
        return df
    except Exception as e:
        print("Could not fetch via ucimlrepo. Provide mushrooms.csv locally.")
        raise e

df = load_mushroom_data()
print(df.shape)
df.head()
```

## 3) Split & Preprocessing
- **One-Hot Encode** all features (handle_unknown='ignore').
- No scaling needed because features are categorical.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

X = df.drop(columns=['class']).copy()
y = df['class'].astype(str).str.strip()  # ensure string labels 'e'/'p'

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

preprocess = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=True), X.columns)],
    remainder='drop'
)

print("Train size:", X_train.shape, " Test size:", X_test.shape)
```

### Optional: Handle Class Imbalance (SMOTE)
Will only apply if SMOTE is installed **and** classes are imbalanced (>60/40).

```python
imbalance_ratio = (y_train.value_counts(normalize=True).max())
APPLY_SMOTE = HAS_SMOTE and (imbalance_ratio > 0.6)

print(f"Imbalance ratio (max class proportion): {imbalance_ratio:.3f}")
print("Apply SMOTE:", APPLY_SMOTE)
```

## 4) Feature Selection
- **Filter (chi2)** with `SelectKBest`
- **Wrapper (RFECV)** with Logistic Regression
- **Embedded** via RandomForest feature importances

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Fit baseline OHE to obtain transformed feature names
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=True)  # <- changed here
ohe.fit(X_train)

# Get names for OHE outputs
try:
    ohe_feature_names = ohe.get_feature_names_out(X.columns)
except AttributeError:
    # Fallback for older sklearn
    ohe_feature_names = ohe.get_feature_names(X.columns)

# Use chi2 to rank top features after OHE
chi2_selector = SelectKBest(score_func=chi2, k=min(30, len(ohe_feature_names)))

# Fit a separate pipeline for ranking
X_train_ohe = ohe.transform(X_train)
scores, pvals = chi2(X_train_ohe, y_train)

chi2_ranks = pd.DataFrame({
    'feature': ohe_feature_names,
    'chi2_score': scores,
    'p_value': pvals
}).sort_values('chi2_score', ascending=False)

chi2_ranks.head(15)
```

```python
rf = RandomForestClassifier(
    n_estimators=300,
    random_state=RANDOM_STATE,
    n_jobs=None
)

rf_pipe = Pipeline(steps=[('pre', preprocess), ('clf', rf)])
rf_pipe.fit(X_train, y_train)

# Extract feature importances on OHE space
X_train_ohe = ohe.transform(X_train)
importances = rf.feature_importances_ if hasattr(rf, 'feature_importances_') else rf_pipe.named_steps['clf'].feature_importances_
imp_df = pd.DataFrame({'feature': ohe_feature_names, 'importance': importances}).sort_values('importance', ascending=False)
imp_df.head(20)
```

## 5) Train 6 Models (Baseline, default hyperparameters)
Models:
- Logistic Regression, Decision Tree, Random Forest, SVC, KNN, BernoulliNB

```python
models = {
    'LogReg': LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    'DecisionTree': DecisionTreeClassifier(random_state=RANDOM_STATE),
    'RandomForest': RandomForestClassifier(n_estimators=300, random_state=RANDOM_STATE),
    'SVC': SVC(probability=True, kernel='rbf', random_state=RANDOM_STATE),
    'KNN': KNeighborsClassifier(),
    'BernoulliNB': BernoulliNB()
}

results = []
fitted_models: Dict[str, Pipeline] = {}

for name, clf in models.items():
    pipe = Pipeline(steps=[('pre', preprocess), ('clf', clf)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    y_proba = pipe.predict_proba(X_test)[:, 1] if hasattr(pipe.named_steps['clf'], 'predict_proba') else None

    acc = accuracy_score(y_test, y_pred)
    # For binary labels like 'e'/'p', map to 0/1 for ROC if needed
    try:
        y_true_bin = (y_test == sorted(y_test.unique())[-1]).astype(int)
        if y_proba is not None:
            roc = roc_auc_score(y_true_bin, y_proba)
        else:
            roc = np.nan
    except Exception:
        roc = np.nan

    results.append({'model': name, 'accuracy': acc, 'roc_auc': roc})
    fitted_models[name] = pipe

    print(f"\n=== {name} ===")
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred, labels=sorted(y_test.unique()))
    print("Confusion Matrix (rows=true, cols=pred):\n", cm)

results_df = pd.DataFrame(results).sort_values(['roc_auc','accuracy'], ascending=False)
results_df.reset_index(drop=True, inplace=True)
results_df
```

## 6) Hyperparameter Tuning (GridSearchCV)
We tune **RandomForest**, **SVC**, and **KNN**.

```python
grids = {
    'RandomForest': (RandomForestClassifier(random_state=RANDOM_STATE), {
        'clf__n_estimators': [200, 400],
        'clf__max_depth': [None, 10, 20],
        'clf__min_samples_split': [2, 5]
    }),
    'SVC': (SVC(probability=True, random_state=RANDOM_STATE), {
        'clf__C': [0.5, 1, 2],
        'clf__gamma': ['scale', 'auto']
    }),
    'KNN': (KNeighborsClassifier(), {
        'clf__n_neighbors': [3, 5, 11],
        'clf__weights': ['uniform', 'distance']
    })
}

tuned_results = []
best_models = {}

for name, (clf, param_grid) in grids.items():
    pipe = Pipeline(steps=[('pre', preprocess), ('clf', clf)])
    gs = GridSearchCV(pipe, param_grid=param_grid, cv=5, scoring='f1', n_jobs=None)
    gs.fit(X_train, y_train)
    best_pipe = gs.best_estimator_
    y_pred = best_pipe.predict(X_test)
    y_proba = best_pipe.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    y_true_bin = (y_test == sorted(y_test.unique())[-1]).astype(int)
    roc = roc_auc_score(y_true_bin, y_proba)

    tuned_results.append({'model': name + ' (tuned)', 'accuracy': acc, 'roc_auc': roc})
    best_models[name] = best_pipe

    print(f"\n=== {name} (tuned) ===")
    print("Best params:", gs.best_params_)
    print(classification_report(y_test, y_pred))

tuned_df = pd.DataFrame(tuned_results).sort_values(['roc_auc','accuracy'], ascending=False)
tuned_df.reset_index(drop=True, inplace=True)
tuned_df
```

## 7) Compare Models (Baseline vs Tuned)

```python
compare_df = pd.concat([results_df, tuned_df], ignore_index=True).sort_values(['roc_auc','accuracy'], ascending=False)
compare_df.reset_index(drop=True, inplace=True)
compare_df
```

## 8) ROC Curves (Top Models)

```python
# Collect up to 4 top models (by ROC AUC) for plotting
plot_candidates = compare_df.dropna(subset=['roc_auc']).head(4)['model'].tolist()

plt.figure()
for name in plot_candidates:
    if name.endswith('(tuned)'):
        base = name.split(' ')[0]
        pipe = best_models.get(base)
    else:
        pipe = fitted_models.get(name)
    if pipe is None:
        continue
    # Probabilities
    y_proba = pipe.predict_proba(X_test)[:, 1]
    y_true_bin = (y_test == sorted(y_test.unique())[-1]).astype(int)
    fpr, tpr, _ = roc_curve(y_true_bin, y_proba)
    plt.plot(fpr, tpr, label=name)

# Diagonal
plt.plot([0,1],[0,1], linestyle='--')
plt.title("ROC Curves â€” Top Models")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
